```python
# Магический шар 8
# Описание проекта: магический шар 8 (шар судьбы) — шуточный способ предсказывать будущее. Программа должна просить пользователя задать некий вопрос, чтобы случайным образом на него ответить.

# Составляющие проекта:

# Целые числа (тип int);
# Переменные;
# Ввод / вывод данных (функции input() и print());
# Условный оператор (if/elif/else);
# Цикл while;
# Бесконечный цикл;
# Операторы break, continue;
# Работа с модулем random для генерации случайных чисел.
# Варианты ответов

# Традиционно шар имеет 
# 20 ответов, которые можно разделить на четыре группы:
# Положительные,	Нерешительно положительные,	Нейтральные,	Отрицательные
```

```python
from random import choice

answers = ['Бесспорно',	'Мне кажется - да',	'Пока неясно, попробуй снова',	'Даже не думай', 'Предрешено',	'Вероятнее всего',	'Спроси позже',	'Мой ответ - нет',	'Никаких сомнений',	'Хорошие перспективы',	'Лучше не рассказывать',	'По моим данным - нет',	'Определённо да',	'Знаки говорят - да',	'Сейчас нельзя предсказать',	'Перспективы не очень хорошие',	'Можешь быть уверен в этом',	'Да',	'Сконцентрируйся и спроси опять',	'Весьма сомнительно']

print('Привет Мир, я магический шар, и я знаю ответ на любой твой вопрос.')
print('Как тебя зовут?')
name = input()
print('Привет,', name)

is_new_question = True
while is_new_question:
    print('Слушаю твой вопрос...')
    print(input())
    print(choice(answers))

    print('Хочешь задать ещё вопрос? (Y/N)')
    yes_no = input()
    while (yes_no != 'N' and yes_no != 'Y'):
        print('Введи Y(да) или N(нет)')
        yes_no = input()

    if yes_no == 'N':
        print('Возвращайся, если возникнут вопросы!')
        is_new_question = False









```
