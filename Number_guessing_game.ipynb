```python
# Проект 1: Игра "Угадайка чисел"
# Часть 1
# Описание: программа генерирует случайное число в диапазоне от 1 до 100 и просит
# пользователя угадать это число. Если догадка пользователя больше случайного числа, 
# то программа должна вывести сообщение 'Слишком много, попробуйте еще раз'. 
# Если догадка меньше случайного числа, то программа должна вывести сообщение 
# 'Слишком мало, попробуйте еще раз'. Если пользователь угадывает число, то программа 
# должна поздравить его и вывести сообщение 'Вы угадали, поздравляем!', 
# а также вывести количество попыток.
# Когда игрок нашел ответ, программа должна предложить сыграть ещё.

# Составляющие проекта:

# Целые числа (тип int);
# Переменные;
# Ввод / вывод данных (функции input() и print());
# Функции
# Условный оператор (if/elif/else);
# Цикл while;
# Бесконечный цикл;
# Операторы break, continue;
# Работа с модулем random для генерации случайных чисел.
# f - строки

from random import randint

# включая границы загаданного числа
left = 1
right = 100

print('Добро пожаловать в числовую угадайку')

# Функция проверки корректности входных данных. 
# Обработка такого рода ситуаций называется "защитой от дурака".
def is_valid(x, left, right):
    return left <= x <= right

# Функция проверки старта новой игры.
def check_YN(symbol):
    if symbol == 'Y':
        return True
    elif symbol == 'N':
        return False
    return -1

# функция с одной игровой итерацией
def game_iteration():
        
    # генерация числа от 1 до 100, включая границы
    x = randint(left, right)
    # print(x)
    counter = 0

    print(f'Вводите, пожалуйста, числа в диапазоне [{left}, {right}]')

    while True:
        new_number = int(input())

        while not is_valid(new_number, left, right):
            print('Давайте все-таки введем целое число от 1 до 100')
            new_number = int(input())

        counter += 1

        if x < new_number:
            print(f'{new_number}: cлишком много, попробуйте еще раз')
        elif x > new_number:
            print(f'{new_number}: cлишком мало, попробуйте еще раз')
        else:
            print(f'{new_number}: вы угадали, поздравляем! Количество попыток: {counter}')
            break

# основная функция, обрабатывающая процесс игры
def game():
    flag_start_new_game = True
    # старт новой итерации игры
    while flag_start_new_game:
            
        game_iteration()

        print('Желаете сыграть снова? (Y/N)')
        flag_start_new_game = check_YN(input())
        while flag_start_new_game == -1:
            print('Введите Y(да) или N(нет)')
            flag_start_new_game = check_YN(input())

    print('Спасибо, что играли в числовую угадайку.')

# запуск игры
game()
```

```python
# Проект 1: Угадайка чисел
#  Часть 2. Определение минимального количества попыток для гарантированного определения заданного числа

# Оптимальный алгоритм угадывания: положим left = 1 и right = 100.
# Называем число, равное middle = (left + right) // 2;
# Если число middle равно задуманному числу, то мы угадали!;
# Если число middle меньше задуманного числа, то положим left = middle + 1 и продолжим алгоритм;
# Если число middle больше задуманного числа, то положим right = middle - 1 и продолжим алгоритм.
# Поскольку на каждой итерации мы отбрасываем половину чисел, то гарантировано угадаем задуманное число за величину, равную 
# log2(n + 1) (двоичный логарифм), округленную до целого в большую сторону. Так, при n=100 получаем 7 попыток.

# Cоставляющие проекта:
# Работа с модулем math.

from math import log2, ceil
n = int(input())
counter_attempts = ceil(log2(n + 1))        
print(counter_attempts)
```
