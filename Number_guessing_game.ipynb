```python
# Проект 1: Угадайка чисел
# Описание: программа генерирует случайное число в диапазоне от 1 до 100
# и просит пользователя угадать это число. Если догадка пользователя больше случайного числа, 
# то программа должна вывести сообщение 'Слишком много, попробуйте еще раз'. 
# Если догадка меньше случайного числа, то программа должна вывести сообщение 
# 'Слишком мало, попробуйте еще раз'. Если пользователь угадывает число, то программа должна 
# поздравить его и вывести сообщение 'Вы угадали, поздравляем!'.

#Составляющие проекта:

#Целые числа (тип int);
#Переменные;
#Ввод / вывод данных (функции input() и print());
#Условный оператор (if/elif/else);
#Цикл while;
#Бесконечный цикл;
#Операторы break, continue;
#Работа с модулем random для генерации случайных чисел.

from random import randint

# генерация числа от 1 до 100, включая границы
x = randint(1, 100) 

message_too_large = 'Слишком много, попробуйте еще раз'
message_too_small = 'Слишком мало, попробуйте еще раз'
message_win = 'Вы угадали, поздравляем'

while True:
    new_number = int(input())
    if x < new_number:
        print(message_too_large)
    elif x > new_number:
        print(message_too_small)
    else:
        print(message_win)
        break
```
